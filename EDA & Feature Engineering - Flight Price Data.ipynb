## Feature Engineering

### Splitting the Data_of_Journey columns into data, Month, Year separtely

```python
dataset['Date']=dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month']=dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year']=dataset['Date_of_Journey'].str.split('/').str[2]
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#Date, Month, Year are still in object so converting the object to int

dataset['Date']=dataset['Date'].astype(int)
dataset['Month']=dataset['Month'].astype(int)
dataset['Year']=dataset['Year'].astype(int)
```

```python
dataset.info()
```

```python
dataset.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#Arrival Timme

dataset['arrival_hour_min']=dataset['Arrival_Time'].str.split(' ').str[0]
```

```python
dataset.head(2)
```

```python
dataset['arrival_hour']=dataset['arrival_hour_min'].str.split(':').str[0]
dataset['arrival_min']=dataset['arrival_hour_min'].str.split(':').str[1]
```

```python
dataset.head(2)
```

```python
dataset.head(2)
```

```python
dataset['Arrival_Time']=dataset['Arrival_Time'].str.split(' ').str[0]
```

```python
dataset.head()
```

```python
dataset['arrival_hour']=dataset['arrival_hour'].astype(int)
dataset['arrival_min']=dataset['arrival_min'].astype(int)
```

```python
dataset.info()
```

```python
dataset.drop('Arrival_Time',axis=1,inplace=True)
```

```python
dataset.head()
```

```python
#Dep_Time

dataset['Dep_hour']=dataset['Dep_Time'].str.split(':').str[0]
dataset['Dep_min']=dataset['Dep_Time'].str.split(':').str[1]
```

```python
dataset.head()
```

```python
dataset.drop('Dep_Time',axis=1,inplace=True)
```

```python
dataset['Dep_hour']=dataset['Dep_hour'].astype(int)
dataset['Dep_min']=dataset['Dep_min'].astype(int)
```
